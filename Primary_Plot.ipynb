```python
import pandas as pd
import numpy as np
import folium
```

```python
data=pd.read_csv("D:/PathPlanning_Data.csv");
features=data[['Median Longitude','Median Latitude']]
features
```

```python
latitudes = data['Median Latitude']
longitudes = data['Median Longitude']

# Create a base map centered around the first entry (or any specific entry you prefer)
if not data.empty:
    center_lat = latitudes.iloc[0]
    center_lon = longitudes.iloc[0]
else:
    center_lat = 0
    center_lon = 0

map = folium.Map(location=[center_lat, center_lon], zoom_start=13)

# Add markers for each location
for lat, lon in zip(latitudes, longitudes):
    folium.Marker(location=[lat, lon]).add_to(map)

map
```
