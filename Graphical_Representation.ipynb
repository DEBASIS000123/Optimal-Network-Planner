```python
import pandas as pd
import numpy as np
import geopy.distance
```

```python
data=pd.read_csv("D:/SupplementaryMaterial.csv")
```

```python
features=data[['Longitude','Latitude']]
features
```

```python
features.replace('\t',pd.NA,inplace=True)
features.replace(' ',pd.NA,inplace=True)
features=features.apply(pd.to_numeric,errors='coerce')
```

```python
#Graphical Representation
coordList = []

for index, row in features.iterrows():
    if not (np.isnan(row['Latitude']) or np.isnan(row['Longitude'])):
        lat = row['Latitude']
        lon = row['Longitude']
        coordList.append([lat, lon])

import folium
from branca.element import Figure
import numpy as np
fig2 = Figure(width=1100, height=450)
m2 = folium.Map(location=[6.437201, 3.502389])

fig2.add_child(m2)



f1 = folium.FeatureGroup("Vehicle 1")
line_1 = folium.vector_layers.PolyLine(coordList, popup='<b>Path of Vehicle_1</b>', tooltip='Vehicle_1', color='purple', weight=2).add_to(f1)
f1.add_to(m2)

folium.LayerControl().add_to(m2)
m2.add_child(folium.Marker(location=[6.43720072, 3.50238888],popup='Start',icon=folium.Icon(color='green')))
m2.add_child(folium.Marker(location=[6.43620545, 3.41596484],popup='End',icon=folium.Icon(color='green')))
```

```python
#Calculating total path
import numpy as np
features.replace('\t',pd.NA,inplace=True)
features.replace(' ',pd.NA,inplace=True)
features=features.apply(pd.to_numeric,errors='coerce')

coordList = []

for index, row in features.iterrows():
    if not (np.isnan(row['Latitude']) or np.isnan(row['Longitude'])):
        lat = row['Latitude']
        lon = row['Longitude']
        coordList.append((lat, lon))


total_distance = 0
for i in range(len(coordList) - 1):
    distance = geopy.distance.geodesic(coordList[i], coordList[i + 1]).kilometers
    total_distance += distance

print("Total path covered:", total_distance, "kilometers")
```
