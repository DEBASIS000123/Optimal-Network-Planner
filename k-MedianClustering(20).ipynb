```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances_argmin_min
from pathlib import Path
```

```python
data=pd.read_csv("../Manhattan_aggoclustered_data100.csv");
data
```

```python

def k_median_clustering(cluster_data, k, max_iterations=100):
    centers_idx = np.random.choice(len(cluster_data), k, replace=False)
    centers = cluster_data.iloc[centers_idx].values
    
    for _ in range(max_iterations):
        labels, _ = pairwise_distances_argmin_min(cluster_data.values, centers, metric='cityblock')
        new_centers = np.array([np.median(cluster_data.values[labels == i], axis=0) for i in range(k)])
        
        if np.all(centers == new_centers):
            break
        
        centers = new_centers
    
    return labels, centers


for i in range(100):
    if i==50:
        continue
    features = data[data['cluster'] == i]
    cluster_data=features[['Latitude','Longitude','Serving RSSI_4G']]
    
    cluster_data.replace('//t', pd.NA, inplace=True)
    cluster_data.replace('', pd.NA, inplace=True)
    cluster_data = features.apply(pd.to_numeric, errors='coerce')
        
    labels, centers = k_median_clustering(cluster_data, 20)
    cluster_data['Cluster_Label'] = labels
    clustered_data = pd.DataFrame()
    clustered_data = pd.concat([clustered_data, cluster_data], ignore_index=True)

    file_path = Path(f'D:/ReClusterData/reclustered{i}_data.csv')
    clustered_data.to_csv(file_path, index=False)
```
