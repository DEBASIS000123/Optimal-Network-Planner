```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#All the Data in SupplementaryMaterial
data=pd.read_csv("D:/SupplementaryMaterial.csv");
data
```

```python
#All the columns in Data file
data.columns
```

```python
features=data[['Longitude','Latitude','Serving RSSI_4G']]
features
```

```python
#Heatmap of features
features.replace('\t',pd.NA,inplace=True)
features.replace(' ',pd.NA,inplace=True)
features=features.apply(pd.to_numeric,errors='coerce')

p1=sns.heatmap(features)
```

```python
#Scatter Plot of features
x=data['Latitude']
x.replace('\t',pd.NA,inplace=True)
x.replace(' ',pd.NA,inplace=True)
x=x.apply(pd.to_numeric,errors='coerce')

c=data['Longitude']
c.replace('\t',pd.NA,inplace=True)
c.replace(' ',pd.NA,inplace=True)
c=c.apply(pd.to_numeric,errors='coerce')

colors=c

y=data['Serving RSRP_4G']
y.replace('\t',pd.NA,inplace=True)
y.replace(' ',pd.NA,inplace=True)
y=y.apply(pd.to_numeric,errors='coerce')
plt.scatter(x,y,s=10,c=colors,cmap=plt.cm.get_cmap("cool",5))
cbar=plt.colorbar(orientation="vertical", extend="both" , pad=0.05,shrink=1,aspect=10, format="%.3f")
cbar.set_label(label="Longitude",size=12)
plt.clim(40,50)
plt.xlabel("Latitude")
plt.ylabel("Serving RSRP_4G")
plt.show()
```

```python
from mpl_toolkits import mplot3d

fig=plt.figure()
ax=plt.axes(projection='3d')
x=data[['Latitude']]
y=data[['Longitude']]
z=data[['Serving RSRP_4G']]

x.replace('\t',pd.NA,inplace=True)
x.replace(' ',pd.NA,inplace=True)
x=x.apply(pd.to_numeric,errors='coerce')

y.replace('\t',pd.NA,inplace=True)
y.replace(' ',pd.NA,inplace=True)
y=y.apply(pd.to_numeric,errors='coerce')

z.replace('\t',pd.NA,inplace=True)
z.replace(' ',pd.NA,inplace=True)
z=z.apply(pd.to_numeric,errors='coerce')


ax.scatter(x,y,z,c=z,cmap='viridis',linewidth=0.8)
```
